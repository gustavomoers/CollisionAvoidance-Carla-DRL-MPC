```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
files = ['run-1709461045-recurrentPPO-50kmh_PPO_0-tag-reward.csv', 'run-1709461045-recurrentPPO-70kmh-transfer_PPO_70_0-tag-reward.csv', 'run-1709461045-recurrentPPO-90kmh-transfer_PPO_90_0-tag-reward.csv']
dfs = []
for file in files:
    dfs.append(pd.read_csv(f'plotting/rew-mpc/{file}', index_col='Step'))
for x in range(1, len(dfs)):
    dfs[x] = dfs[x].loc[dfs[x-1].index[-1]:]




df = pd.concat(dfs)
```

```python
from scipy import stats
import numpy as np

df = df[np.abs(stats.zscore(df['Value'])) < 3]
```

```python
from bokeh.io import output_notebook

output_notebook()
```

```python

from bokeh.plotting import figure, show
from scipy.interpolate import make_interp_spline
import numpy as np
 
p = figure(
            width=1110, 
            height=350, 
            background_fill_color="#fafafa",
            title='Hybrid Approach - Reward per Step',
            x_axis_label='Steps',
            y_axis_label='Reward',)
            # y_range=(-0.61, 0.61))

# generate some values
smooth = df.ewm(alpha=(1 - 0.99)).mean()
x = smooth.index
y = smooth['Value']


# add a line renderer and legend to the plot
p.line(x, y, line_color='red', line_width=2)
p.below[0].formatter.use_scientific = False
# p.legend.location = "top_right"


# show the results
show(p)
```

```python
from bokeh.io import export_svgs

# option 1
p.output_backend = "svg"

# option 2
export_svgs(p, filename="rew-step-mpc.svg")
```
